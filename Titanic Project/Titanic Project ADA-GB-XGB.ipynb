# WELCOME!

**Overview**

The data has been split into two groups:

- training set (train.csv)
- test set (test.csv)

**The training set** should be used to build your machine learning models. For the training set, we provide the outcome (also known as the ground truth) for each passenger. Your model will be based on features like passengers gender and class. You can also use feature engineering to create new features.

**The test set** should be used to see how well your model performs on unseen data. For the test set, we do not provide the ground truth for each passenger. It is your job to predict these outcomes. For each passenger in the test set, use the model you trained to predict whether or not they survived the sinking of the Titanic.

We also include gender_submission.csv, a set of predictions that assume all and only female passengers survive, as an example of what a submission file should look like.

**Data Dictionary**



**Variable Notes**

pclass: A proxy for socio-economic status (SES)

1st = Upper

2nd = Middle

3rd = Lower

age: Age is fractional if less than 1. If the age is estimated, is it in the form of xx.5

sibsp: The dataset defines family relations in this way...

Sibling = brother, sister, stepbrother, stepsister

Spouse = husband, wife (mistresses and fiancs were ignored)

parch: The dataset defines family relations in this way...

Parent = mother, father

Child = daughter, son, stepdaughter, stepson

Some children travelled only with a nanny, therefore parch=0 for them.

PassengerId: Yolcunun benzersiz kimlik numaras.

Survived: Yolcunun hayatta kalp kalmadn belirtir. (0 = Hayatta kalmad, 1 = Hayatta kald)

Pclass: Yolcunun bilet snf. (1 = Birinci Snf, 2 = kinci Snf, 3 = nc Snf)

Name: Yolcunun tam ad.

Sex: Yolcunun cinsiyeti. (male = Erkek, female = Kadn)

Age: Yolcunun ya. (Yarm yalar da dahil olabilir, rn. 22.5)

SibSp: Yolcunun gemideki e veya karde says.

Parch: Yolcunun gemideki ebeveyn veya ocuk says.

Ticket: Yolcunun bilet numaras.

Fare: Yolcunun dedii bilet creti.

Cabin: Yolcunun kald kabin numaras (Baz kaytlar eksik olabilir).

Embarked: Yolcunun gemiye bindii liman belirtir. (C = Cherbourg, Q = Queenstown, S = Southampton)

# Encoding

Tree based modellerle ilgili yaplan akademik bir almada;

1. Ordinal encoder scorlarnn one hot encoder scorlarna gre ortalama %1, %2 civarlarnda daha iyi olduu,

2. eitim esnasnda ordinal encodern, one hot encodera gre 2,5 kat daha hzl alt,

3. Doru feature importance elde edebilmek iin Ordinal encoder kullanlmas gerektii grlmtr.

Tree based modellerde amacmz sadece skor almak ise hem one hot encoder hem de ordinal encoder kullanabiliriz. 

Ancak, yukarda bahsettiimiz avantajlarndan dolay tree based modellerde ordinal encoder kullanacaz.

Eer amacnz sadece skor almaksa ordinal encoder veya one hot encoder kullanabileceimiz sylemitik. Ancak kesinlikle

get_dummies fonksiyonunu kullanarak gzlemleri dummies featurlara (one hot encoder) dntrmyoruz. nk get_dummies 

fonksiyonu data leakage sorununa yol aar. get_dummies yerine her zaman one hot encoder kullanmamz gerektiini unutmuyoruz.

```python
cols=['Sex', 'Embarked']
enc1=OrdinalEncoder(categories=[['female', 'male']])
enc2=OrdinalEncoder(categories=[['S','C','Q','U']])
enc = [enc1, enc2]
for i, k in zip(cols, enc):
    train[i]=k.fit_transform(train[[i]])
    test[i]=k.transform(test[[i]])
```

# Compare The Models

```python
compare = pd.DataFrame({"Model": ["AdaBoost", "GradientBoost", "XGBoost"],
                        "Accuracy": [ada_accuracy_test, grb_accuracy_test, xgb_accuracy_test],
                        "F1": [ada_f1_test, grb_f1_test, xgb_f1_test],
                        "ROC_AUC": [ada_auc, grb_auc, xgb_auc],
                        "PRC": [ada_prc, grb_prc, xgb_prc]})

def labels(ax):
    for p in ax.patches:
        width = p.get_width()                        # get bar length
        ax.text(width,                               # set the text at 1 unit right of the bar
                p.get_y() + p.get_height() / 2,      # get Y coordinate + X coordinate / 2
                '{:1.3f}'.format(width),             # set variable to display, 2 decimals
                ha = 'left',                         # horizontal alignment
                va = 'center')                       # vertical alignment
    
plt.figure(figsize=(16,16))

plt.subplot(411)
compare = compare.sort_values(by="F1", ascending=False)
ax=sns.barplot(x="F1", y="Model", data=compare, palette="Blues_d")
labels(ax)

plt.subplot(412)
compare = compare.sort_values(by="Accuracy", ascending=False)
ax=sns.barplot(x="Accuracy", y="Model", data=compare, palette="Blues_d")
labels(ax)

plt.subplot(413)
compare = compare.sort_values(by="ROC_AUC", ascending=False)
ax=sns.barplot(x="ROC_AUC", y="Model", data=compare, palette="Blues_d")
labels(ax)

plt.subplot(414)
compare = compare.sort_values(by="PRC", ascending=False)
ax=sns.barplot(x="PRC", y="Model", data=compare, palette="Blues_d")
labels(ax)

plt.show()
```
